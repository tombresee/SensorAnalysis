````python
df = pd.read_parquet('single_node_all_data_huge.parquet', engine='fastparquet')

# this ends up being a 270MB single node location all data file
# representing 82.4 millions rows of data...

df.value_hrf = df.value_hrf.astype("float")
````

## Examining some random sensor data

````python
single_sensor_single_parameter = df[df.sensor=='sht25'].copy()
single_sensor_single_parameter = df[df.parameter=='humidity']
len(single_sensor_single_parameter) # 10 278 158, 10M rows...
````

````python
minidf = df.loc['2018-01-01':'2019-10-31'].copy()
len(minidf)
````

````python
minidf.head(10) # every 25s, it adds up fast =
````

##### Lets only look at temperature sensors...

````python
tdf = minidf[minidf.parameter == 'temperature'].copy()
tdf.sensor = tdf.sensor.astype(object)
tdf.parameter = tdf.parameter.astype(object)
tdf = tdf[tdf.sensor == 'tsys01']
plot_data = pd.DataFrame(tdf['value_hrf'], columns=['value_hrf'], index=tdf.index)
plot_data.head(10)
````

````python
curve = hv.Curve(plot_data, 'timestamp', ('value_hrf', 'Temperature (C)'))
````

````python
# REAL-TIME PLOT AVAILABLE IN HTML BELOW ! 
````

````python
hv.extension('bokeh')
tgt = curve.relabel('Temperature Analysis').opts(width=800, color='blue', 
                                                 fontsize={'title': 10, 'labels': 8, 'xticks': 6, 'yticks': 6}, 
                                                 line_width = .5, labelled=['y'], toolbar='disable')

src = curve.opts(width=800, height=150, yaxis=None, line_width = .3, color='grey', default_tools=[], 
                 fontsize={'title': 9, 'labels': 10, 'xticks': 6, 'yticks': 12})

RangeToolLink(src, tgt)

layout = (tgt + src).cols(1)
layout = layout.opts(opts.Layout(shared_axes=False, merge_tools=False))
layout 
````
